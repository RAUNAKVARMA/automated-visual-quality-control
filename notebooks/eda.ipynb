# MVTec AD — exploratory data analysis

After downloading MVTec AD (see `scripts/download_dataset.md`), set `MVTEC_ROOT` below and run the cells to preview a few training images.

```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

MVTEC_ROOT = Path("../data/mvtec")
CATEGORY = "bottle"
train_good = MVTEC_ROOT / CATEGORY / "train" / "good"
paths = sorted(train_good.glob("*.png"))[:6]
if not paths:
    raise SystemExit(f"No images in {train_good} — download MVTec AD first.")
fig, axes = plt.subplots(2, 3, figsize=(10, 6))
for ax, p in zip(axes.ravel(), paths):
    ax.imshow(Image.open(p))
    ax.set_title(p.name, fontsize=8)
    ax.axis("off")
plt.suptitle(f"{CATEGORY} — train/good samples")
plt.tight_layout()
```
